## Función 14: `resumen_calificaciones(calificaciones)`

**Archivo:** `src/calificaciones.py`

Regresa un diccionario con el resumen general de una lista de calificaciones.

Debe incluir las siguientes claves:

- `"total"`
- `"promedio"`
- `"moda"`
- `"mediana"`
- `"maxima"`
- `"minima"`
- `"porcentaje_aprobados"`
- `"porcentaje_reprobados"`
- `"distribucion"`

En `"distribucion"` se debe colocar el resultado de `frecuencia_calificaciones(calificaciones)`.

**Entrada:**

```python
[65, 65, 70, 80, 80, 80, 90, 100]
```

**Salida esperada:**

```python
{
    "total": 8,
    "promedio": 78.75,
    "moda": 80,
    "mediana": 80.0,
    "maxima": 100,
    "minima": 65,
    "porcentaje_aprobados": 75.0,
    "porcentaje_reprobados": 25.0,
    "distribucion": {
        65: 2,
        70: 1,
        80: 3,
        90: 1,
        100: 1
    }
}
```

**Caso especial:**

Si la lista está vacía, debe regresar un diccionario con:

```python
{
    "total": 0,
    "promedio": None,
    "moda": None,
    "mediana": None,
    "maxima": None,
    "minima": None,
    "porcentaje_aprobados": 0.0,
    "porcentaje_reprobados": 0.0,
    "distribucion": {}
}
```

**Test:**

```bash
python -m unittest tests/test_14_resumen_calificaciones.py
```

## Función 7: `clasificar_calificacion(calificacion)`

**Archivo:** `src/calificaciones.py`

Clasifica una calificación usando estos criterios:

| Rango | Clasificación |
|---|---|
| 90 a 100 | `"Excelente"` |
| 80 a 89 | `"Bueno"` |
| 70 a 79 | `"Regular"` |
| Menor que 70 | `"Reprobado"` |

**Entradas y salidas esperadas:**

```python
95 -> "Excelente"
85 -> "Bueno"
75 -> "Regular"
60 -> "Reprobado"
```

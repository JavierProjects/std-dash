# Reportes generales

En este módulo se integran los resultados de todo el sistema académico.

## Función 28: `reporte_general(calificaciones, asistencias, respuestas)`

**Archivo:** `src/reportes.py`

Integra resultados de calificaciones, asistencias y encuestas en un solo diccionario.

Para esta función trabajaremos con tres estructuras de datos relacionadas:

* una lista de calificaciones;
* un diccionario de asistencias;
* una lista de respuestas.

En este ejemplo, las tres estructuras representan información de **8 estudiantes**, por lo que deben tener la misma cantidad de elementos.

Es decir:

* 8 calificaciones;
* 8 alumnos en el diccionario de asistencias;
* 8 respuestas de encuesta.

La función debe incluir las siguientes claves:

* `"promedio_calificaciones"`
* `"porcentaje_aprobados"`
* `"promedio_asistencia"`
* `"respuesta_mas_comun"`

**Entrada:**

```python
calificaciones = [65, 65, 70, 80, 80, 80, 90, 100]

asistencias = {
    "Ana": [1, 1, 1, 1],
    "Luis": [1, 0, 0, 1],
    "Hugo": [1, 1, 1, 1],
    "Monse": [1, 0, 0, 1],
    "Liz": [1, 1, 1, 1],
    "Maria": [1, 0, 0, 1],
    "Pedro": [1, 1, 1, 1],
    "Marco": [1, 0, 0, 1]
}

respuestas = [
    "Python", "Java", "Python", "Python",
    "Java", "Python", "Python", "Java"
]
```

**Salida esperada:**

```python
{
    "promedio_calificaciones": 78.75,
    "porcentaje_aprobados": 75.0,
    "promedio_asistencia": 75.0,
    "respuesta_mas_comun": "Python"
}
```

**Explicación:**

El promedio de calificaciones es:

```python
78.75
```

El porcentaje de aprobados es:

```python
75.0
```

porque 6 de las 8 calificaciones son mayores o iguales a 70.

El promedio de asistencia es:

```python
75.0
```

porque hay alumnos con 100% de asistencia y alumnos con 50% de asistencia.

La respuesta más común es:

```python
"Python"
```

porque aparece 5 veces en la lista de respuestas.


## Sugerencia de integración

Esta función puede apoyarse en funciones ya construidas en otros módulos:

- `promedio(calificaciones)`
- `porcentaje_aprobados(calificaciones)`
- `promedio_asistencia_grupo(asistencias)`

El objetivo de esta función es practicar cómo una función más grande puede construirse reutilizando funciones pequeñas.

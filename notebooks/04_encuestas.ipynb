# Encuestas

En este módulo se analizan respuestas categóricas usando listas, diccionarios y frecuencias.

## Función 22: `contar_respuestas(respuestas)`

**Archivo:** `src/encuestas.py`

Cuenta cuántas respuestas hay en una lista.

**Entrada:**

```python
["Python", "Java", "Python"]
```

**Salida esperada:**

```python
3
```

**Caso especial:**

```python
[] -> 0
```

**Test:**

```bash
python -m unittest tests/test_22_contar_respuestas.py
```

## Función 23: `obtener_opciones(respuestas)`

**Archivo:** `src/encuestas.py`

Regresa una lista con las opciones únicas.

Debe conservar el orden en el que aparecen por primera vez.

**Entrada:**

```python
["Python", "Java", "Python", "C++", "Java"]
```

**Salida esperada:**

```python
["Python", "Java", "C++"]
```

**Test:**

```bash
python -m unittest tests/test_23_obtener_opciones.py
```

## Función 26: `porcentaje_respuesta(respuestas, opcion)`

**Archivo:** `src/encuestas.py`

Calcula el porcentaje que representa una opción específica.

El resultado debe estar en escala de `0` a `100`.

**Entrada:**

```python
["Python", "Java", "Python"], "Python"
```

**Salida esperada:**

```python
66.66666666666666
```

**Otros casos:**

```python
["Python", "Java", "Python"], "C++" -> 0.0
[], "Python" -> 0.0
```

**Test:**

```bash
python -m unittest tests/test_26_porcentaje_respuesta.py
```

## Función 27: `resumen_encuesta(respuestas)`

**Archivo:** `src/encuestas.py`

Regresa un diccionario con resumen de la encuesta.

Debe incluir:

- `"total"`
- `"opciones"`
- `"frecuencias"`
- `"mas_comun"`

**Entrada:**

```python
["Python", "Java", "Python", "C++", "Python"]
```

**Salida esperada:**

```python
{
    "total": 5,
    "opciones": ["Python", "Java", "C++"],
    "frecuencias": {
        "Python": 3,
        "Java": 1,
        "C++": 1
    },
    "mas_comun": "Python"
}
```

**Caso especial:**

```python
{
    "total": 0,
    "opciones": [],
    "frecuencias": {},
    "mas_comun": None
}
```

**Test:**

```bash
python -m unittest tests/test_27_resumen_encuesta.py
```
